# Fix dwell time and central frequency of Toi's basis


Tois basis has a numerical rounding error in its dwell time and a spectrometer frequency that does not exactly match the one read from the bruker files. Here, These values are manually fixed the FSL MRS basis json files.

```python
for basis_element in os.listdir('./fsl_mrs'):
    with open('./fsl_mrs/'+basis_element) as f:
        basis_json = json.load(f)
    basis_json['basis']['basis_dwell'] = 0.0002
    basis_json['basis']['basis_centre'] = 400.263989952738
    with open('./fsl_mrs_fixed/'+basis_element, 'w') as f:
        json.dump(basis_json, f)
```
